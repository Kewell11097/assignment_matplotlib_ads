```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset into pandas dataframe
df = pd.read_csv('jumta.csv')
```

```python
#exploring the data
df.head()
```

```python
print(df.shape)
print(df.isna().sum())
print(df.info())
```

```python
#line plot showing monthly revenue 
x_line = df['Month']
y_line = df['Revenue']

plt.figure(figsize=(15,10))
plt.plot(x_line, y_line, color='green', label='Monthly Revenue', linewidth=2)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue for 2 year period')
plt.grid()
plt.xticks(rotation=60)
plt.legend()
plt.savefig('line_plot.png')
plt.show()
```

```python
#bar plot showing total orders per month
x_bar = df['Month']
y_bar = df['Orders']

plt.figure(figsize=(15,10))
plt.bar(x_bar, y_bar, color='green', label='Orders', edgecolor='gray')
plt.xlabel('Month')
plt.ylabel('Orders')
plt.title('Orders per month')
plt.xticks(rotation=60)
plt.grid(axis='y')
plt.legend()
plt.savefig('bar_plot.png')
plt.show()
```

```python
#scatter plot showing relationship between orders and average order value
x_scatter = df['Orders']
y_scatter = df['Average_Order_Value']

plt.figure(figsize=(15,10))
plt.scatter(x_scatter, y_scatter, color='green', label='Average Value')
plt.xlabel('Orders')
plt.ylabel('Average Order Value')
plt.title('Orders vs Average Order Value')
plt.grid()
plt.legend()
plt.savefig('scatter_plot.png')
plt.show()
```

```python
#create subplot
fig, (ax_line, ax_bar, ax_scatter) = plt.subplots(nrows=3, ncols=1, figsize=(15, 20))

#line chart
ax_line.plot(x_line, y_line, color='green', label='Monthly Revenue', linewidth=2)
ax_line.set_xlabel('Month')
ax_line.set_ylabel('Revenue')
ax_line.set_title('Monthly Revenue for 2 year period')
ax_line.grid()
ax_line.tick_params(axis='x', rotation=60)
ax_line.legend()

#bar chart
ax_bar.bar(x_bar, y_bar, color='green', label='Orders', edgecolor='gray')
ax_bar.set_xlabel('Month')
ax_bar.set_ylabel('Orders')
ax_bar.set_title('Orders per month')
ax_bar.tick_params(axis='x', rotation=60)
ax_bar.grid(axis='y')
ax_bar.legend()

#scatter chart
ax_scatter.scatter(x_scatter, y_scatter, color='green', label='Average Value')
ax_scatter.set_xlabel('Orders')
ax_scatter.set_ylabel('Average Order Value')
ax_scatter.set_title('Orders vs Average Order Value')
ax_scatter.grid()
ax_scatter.legend()

plt.tight_layout()
plt.savefig('Subpplot_all_visualizations', dpi=300)
plt.show
```
